```python
import json
import os
```

```python
def inspect_coco_json(file_path):
    print(f"--- Loading database: {file_path} ---\n")
    
    # Open and load the JSON file into memory
    with open(file_path, "r") as file:
        data = json.load(file)
        
    # 1. See the main categories (Top-Level Keys)
    print("1. TOP-LEVEL KEYS IN THE DATABASE:")
    print(list(data.keys()))
    print("\n" + "="*50 + "\n")
    
    # 2. Look inside each key to see what kind of data it holds
    print("2. DATA STRUCTURE BREAKDOWN:")
    for key in data.keys():
        section_data = data[key]

        # If it's a dictionary, look inside and print every sub-key
        if isinstance(section_data, dict):
            print(f"- '{key}': is a Dictionary with {len(section_data.keys())} sub-keys.")
            print("  [Sub-keys breakdown:]")
            
            # Loop through all the sub-keys and print their names and data types
            """for sub_key, sub_value in section_data.items():
                print(f"    -> {sub_key}")
            print("\n")"""
        
        # If it's a list, tell us how many items are in it and show the first one
        elif isinstance(section_data, list):
            print(f"- '{key}': is a List containing {len(section_data)} items.")
            if len(section_data) > 0:
                print(f"  [Sample of the first item in '{key}':]")
                sample = json.dumps(section_data[0], indent=4)
                print("  " + sample + " ...\n")
            
        
        # If it's just a simple text/number value
        else:
            print(f"- '{key}': contains a simple {type(section_data).__name__} value.\n")
```

```python
def inspect_image(file_path):
    # Open and load the JSON file into memory
    with open(file_path, "r") as file:
        data = json.load(file)

    # Look for the 'images' key in the data
    if 'images' in data:
        images = data['images']
        print(f"Found 'images' key with {len(images)} entries.\n")
        
        # Show the first image entry as a sample
        if len(images) > 0:
            print("Sample image entry:")
            sample_image = json.dumps(images[:10], indent=4)
            print(sample_image)
    else:
        print("No 'images' key found in the database.")
```

```python
if __name__ == "__main__":
    base_dir = os.path.dirname(os.path.abspath(''))

    annotation_dir = os.path.join(base_dir, "IMAGE_PROCESSING", "coco_annotations", "annotations", "instances_val2017.json")
    
    try:
        inspect_image(annotation_dir)
    except FileNotFoundError:
        print(f"Error: Could not find '{annotation_dir}'. Make sure it is in the same folder as this script!")
```
